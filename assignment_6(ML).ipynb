```python
df=pd.read_csv("/content/Market_Basket.csv",header=None)
df
```

```python
len(df)
```

```python
transactions = []

for i in range(len(df)):
    transaction = df.iloc[i].dropna().tolist()
    transactions.append(transaction)
```

```python
transactions
```

```python
from apyori import apriori
```

```python
rule = apriori(transactions, min_support=0.1, min_confidence=0.1, min_length=3)
```

```python
rule
```

```python
result = list(rule)
result
```

```python
len(result)
```

```python
result_list=[]
for items in result:
    temp=[]
    temp.append(items[0])
    temp.append(items[1])
    temp.append(items[2][0][2])
    temp.append(items[2][0][3])
    result_list.append(temp)

```

```python
result_data = pd.DataFrame(result_list, columns=['Rule','Support','Confidence','Lift'])
result_data
```
